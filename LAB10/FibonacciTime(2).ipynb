```python
from time import time
```

```python
def fib(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

```python
def Fibonacci_Iterativo(n):
    t0 = time()
    k = fib(n)
    tf = time()
    print("  n = ",n,"\n fib(",n,") = ",k,"\n time: ", tf - t0, " seconds process time")
```

```python
print("\tFibonacci con Iterativo\n")
for i in range(501):
    Fibonacci_Iterativo(i)
```
