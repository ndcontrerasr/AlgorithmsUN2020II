```python
# A Python program to print all 
# permutations using library function 
from itertools import permutations 
import numpy as np

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
def bubbleSort(nlist,e):
    a=0
    b=1
    c=a
    while(b<len(nlist)):
        if(nlist[a]>nlist[b]):
            e+=1
            c=b
            d=nlist[a]
            nlist[a]=nlist[b]
            nlist[b]=d
        a+=1
        b+=1
#     print(nlist[c:])
#     print(nlist[:c])
    mat = nlist[c:]
    mas=[]
    if(len(nlist[:c])>0):
        e, mas = bubbleSort(nlist[:c],e)
        
        mat = mas+mat
    return e, mat
```

```python
a = int(input("Tamaño de la lista: "))
mat = []
for i in range(a):
    mat.append(i+1)
# print(" | ",bubbleSort(list(mat)))
# Get all permutations
perm = permutations(mat)
mar = []
e=0
for i in list(perm):
    mai = []
    e=0
    e, mai = bubbleSort(list(i),e)
    mar.append(e)
#     print(list(i)," | ",mai, " | ", e)
print("Finalizo")
#print(mat)
```
